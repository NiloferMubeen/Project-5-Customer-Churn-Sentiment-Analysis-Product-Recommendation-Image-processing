```python
columns = ['device_browser', 'device_operatingSystem', 'device_isMobile','device_deviceCategory','geoNetwork_latitude','geoNetwork_longitude',
           'youtube','earliest_source', 'latest_source', 'earliest_medium', 'latest_medium','earliest_keyword', 'latest_keyword', 'products_array',
           'earliest_isTrueDirect','latest_isTrueDirect','target_date','last_visitId','latest_visit_id','visitId_threshold','earliest_visit_id']
data1 = data.drop(columns,axis =1)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data1['channelGrouping'] = encoder.fit_transform(data1['channelGrouping'])
data1['geoNetwork_region'] = encoder.fit_transform(data1['geoNetwork_region'])
```

```python
data1.dtypes
```

```python
data1['sessionQualityDim']
```

```python
from sklearn.ensemble import RandomForestClassifier
x = data1.drop('has_converted',axis=1) #data,dropping target  
y = data1['has_converted'] #target
rf = RandomForestClassifier(n_estimators=400)
rf.fit(x,y)
```

```python
df = pd.DataFrame({"column":x.columns,
             "scores":rf.feature_importances_
             }).sort_values('scores',ascending =False).head(5)['column']
list(df)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
x = abs(data1.drop('has_converted',axis=1))  # Data
y = data1['has_converted']  # Target

selection = SelectKBest(chi2, k=10)
data_select = selection.fit_transform(x,y)
```

```python
best = pd.DataFrame({
    "columns":x.columns,
    "chi-sq-value": selection.scores_
}).sort_values('chi-sq-value',ascending=False).head(8)
print(best)
list(best['columns'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(with_mean= True)
to_scale = data1[['num_interactions','sessionQualityDim','time_on_site','count_hit','avg_session_time_page']]
scaler.fit(to_scale)
scaled_features= scaler.transform(to_scale)
y = data1['has_converted'] 
```

```python
data_feat = pd.DataFrame(scaled_features, columns=['num_interactions','sessionQualityDim','time_on_site','count_hit','avg_session_time_page'])
data_feat.head()
```

Logistic Regression

```python
from sklearn.model_selection import train_test_split
X = data_feat
y = data1['has_converted']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logm = LogisticRegression()
logm.fit(X_train,y_train)
```

```python
train_pred = logm.predict(X_train)
```

```python
test_pred = logm.predict(X_test)
```

Train performance Metrics

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print("LOGISTIC REGRESSION","\n","----------------------")
print("Train_data Classification report:","\n",classification_report(y_train,train_pred))
print("-------------------------------------------------------------------------")
print("Train data confusion matrix:","\n",confusion_matrix(y_train,train_pred))
```

Test Performance Metrics

```python
print("LOGISTIC REGRESSION","\n","----------------------")
print("Test_data Classification report:","\n",classification_report(y_test,test_pred))
print("-------------------------------------------------------------------------")
print("Test data confusion matrix:","\n",confusion_matrix(y_test,test_pred))
```

K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train,y_train)
```

```python
knn_train_pred = knn.predict(X_train)
knn_test_pred = knn.predict(X_test)
```

```python
print("K Nearest Neighbours","\n","------------------------")
print("Train_data Classification report:","\n",classification_report(y_train,knn_train_pred))
print("-------------------------------------------------------------------------")
print("Train data confusion matrix:","\n",confusion_matrix(y_train,knn_train_pred))
```

```python
print("K Nearest Neighbours","\n","------------------------")
print("Test_data Classification report:","\n",classification_report(y_test,knn_test_pred))
print("-------------------------------------------------------------------------")
print("Test data confusion matrix:","\n",confusion_matrix(y_test,knn_test_pred))
```

Random Forest

```python
X = data1[['transactionRevenue','num_interactions','sessionQualityDim','time_on_site','count_hit','avg_session_time_page']]
y = data1['has_converted']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
rfc = RandomForestClassifier(n_estimators=350)
rfc.fit(X_train.values,y_train.values)
```

```python
rf_train_pred = rfc.predict(X_train.values)
rf_test_pred = rfc.predict(X_test.values)
```

```python
print("RANDOM FOREST","\n","-------------------")
print("Train_data Classification report:","\n",classification_report(y_train,rf_train_pred))
print("-------------------------------------------------------------------------")
print("Train data confusion matrix:","\n",confusion_matrix(y_train,rf_train_pred))
```

```python
print("RANDOM FOREST","\n","-------------------")
print("Test_data Classification report:","\n",classification_report(y_test,rf_test_pred))
print("-------------------------------------------------------------------------")
print("Test data confusion matrix:","\n",confusion_matrix(y_test,rf_test_pred))
```
